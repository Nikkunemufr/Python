```python
X = pd.read_csv('X.csv', header=None, sep=" ")
y = np.array(pd.read_csv('y.csv', header=None, sep=" "))
y.reshape(y.shape[0],)
X_test = pd.read_csv('X_test.csv', header=None, sep=" ")
```

KNN
---

```python
tuned_parameters = {
    "n_neighbors": [3,5,11,19],
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan"]
}

clf_grid_knn = GridSearchCV(KNeighborsClassifier(),
                        tuned_parameters,
                        verbose = 1,
                        cv = 5,
                        n_jobs = -1
                       )
clf_grid_knn.fit(X, y)

print('\033[34mMeilleur jeu de paramètre trouvé parmi les paramètres proposés:\033[0m')
print(clf_grid_knn.best_estimator_,'\n')
print("\033[34mGrille des scores par combinaison de paramètres testés:\033[0m")
    
elements = clf_grid_knn.cv_results_
for i in np.argsort(elements['mean_test_score']):
    print("%0.3f (+/-%0.03f) for %r"% (elements['mean_test_score'][i],
                                       elements['std_test_score'][i],
                                       elements['params'][i]))
```

SVM
---

```python
tuned_parameters = [{'kernel': ['rbf'],
                     'gamma': [1e-1, 1e-2, 1e-3, 1e-4],
                     'C': [1e-2,1e-1,1, 10, 100, 1000]},
                    {'kernel': ['linear'],
                     'C': [1e-2,1e-1,1, 10, 100, 1000]}]

clf_grid_svm = GridSearchCV(svm.SVC(probability=True), 
                        tuned_parameters,
                        cv=5,
                        n_jobs = -1,
                        scoring='accuracy'
                       )
clf_grid_svm.fit(X, y)

print('\033[34mMeilleur jeu de paramètre trouvé parmi les paramètres proposés:\033[0m')
print(clf_grid_svm.best_estimator_,'\n')
print("\033[34mGrille des scores par combinaison de paramètres testés:\033[0m")
    
elements = clf_grid_svm.cv_results_
for i in np.argsort(elements['mean_test_score']):
    print("%0.3f (+/-%0.03f) for %r"% (elements['mean_test_score'][i],
                                       elements['std_test_score'][i],
                                       elements['params'][i]))
```

Regression logistique
---

```python
tuned_parameters = {
    "C":np.logspace(-3,3,7),
    "penalty":["l1","l2"]
}
clf_grid_logi = GridSearchCV(LogisticRegression(),
                        tuned_parameters,
                        cv = 5,
                        n_jobs = -1,
                       )
clf_grid_logi.fit(X, y)

print('\033[34mMeilleur jeu de paramètre trouvé parmi les paramètres proposés:\033[0m')
print(clf_grid_logi.best_estimator_,'\n')
print("\033[34mGrille des scores par combinaison de paramètres testés:\033[0m")
    
elements = clf_grid_logi.cv_results_
for i in np.argsort(elements['mean_test_score']):
    print("%0.3f (+/-%0.03f) for %r"% (elements['mean_test_score'][i],
                                       elements['std_test_score'][i],
                                       elements['params'][i]))
```

MLP Classifier
---

```python
tuned_parameters = {
    'hidden_layer_sizes': [(50,50,50), (50,100,50), (100,)],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant','adaptive'],
}
clf_grid_mlp = GridSearchCV(MLPClassifier(),
                        tuned_parameters,
                        cv = 5,
                        n_jobs = -1,
                       )
clf_grid_mlp.fit(X, y)

print('\033[34mMeilleur jeu de paramètre trouvé parmi les paramètres proposés:\033[0m')
print(clf_grid_mlp.best_estimator_,'\n')
print("\033[34mGrille des scores par combinaison de paramètres testés:\033[0m")
    
elements = clf_grid_mlp.cv_results_
for i in np.argsort(elements['mean_test_score']):
    print("%0.3f (+/-%0.03f) for %r"% (elements['mean_test_score'][i],
                                       elements['std_test_score'][i],
                                       elements['params'][i]))
```

Best classifier
---

1.1 - Accuracy
---

```python
y_pred_knn = clf_grid_knn.best_estimator_.predict(X)

y_pred_svm = clf_grid_svm.best_estimator_.predict(X)

y_pred_logi = clf_grid_logi.best_estimator_.predict(X)

y_pred_mlp = clf_grid_mlp.best_estimator_.predict(X)


acc_knn = accuracy_score(y, y_pred_knn)
print("Accuracy d un classifieur KNN:", acc_knn)

acc_svm = accuracy_score(y, y_pred_svm)
print("Accuracy d un classifieur SVM:", acc_svm)

acc_logi = accuracy_score(y, y_pred_logi)
print("Accuracy d un classifieur Logistic Regression:", acc_logi)

acc_mlp = accuracy_score(y, y_pred_mlp)
print("Accuracy d un classifieur MLP:", acc_mlp)
```
